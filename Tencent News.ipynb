# 下面开始爬取新闻评论及评论用户的个人信息，但目前存在一些问题（多个url的新闻评论被合成到一个list里了）

```python
commentlist = []
user_idlist = []
user_namelist = []
user_attrslist = []
timelist = []

for i in range(len(cmdlist)):
    headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'
    }
    if cmdlist[i]=='无评论':
        commentlist.append('')
        user_idlist.append('')
        user_namelist.append('')
        timelist.append('')
    else:
        cmtu0 = r'http://coral.qq.com/article/'+cmdlist[i]+r'/comment/v2?callback=_article'+cmdlist[i]+r'commentv2&orinum=30&oriorder=o&pageflag=1&cursor='
        cmtu1 =r'&scorecursor=0&orirepnum=2&reporder=o&reppageflag=1&source=1&_=1540911539487'
        cmtu = cmtu0+'0'+cmtu1
        q = requests.get(cmtu,timeout = 50,headers=headers)
        q.raise_for_status()
        cmdjs = re.search('\\((.+)\\)',q.text)
        js = json.loads(cmdjs.group(1))
        #print(js)
        oriretnum = js['data']['oriretnum']
        for k in range(oriretnum):
            comment = js['data']['oriCommList'][k]['content']
            commentlist.append(comment)
            user_id = js['data']['oriCommList'][k]['userid']
            user_idlist.append(user_id)
            #print(type(user_id))
            timestamp = js['data']['oriCommList'][k]['time']
            time = datetime.datetime.fromtimestamp(int(timestamp)).strftime('%Y-%m-%d %H:%M:%S')
            timelist.append(time)
            user_name = js['data']['userList'][str(user_id)]['nick']
            user_namelist.append(user_name)
            user_attrs = js['data']['userList'][str(user_id)]['region']
            user_attrslist.append(user_attrs)
    
#print(commentlist)
#print(user_idlist)
#print(timelist)
#print(user_namelist)
#print(user_attrslist)
```

```python
%%time
testcom1 = r'http://coral.qq.com/article/3280942323/comment/v2?callback=_article3280942323commentv2&orinum=30&oriorder=o&pageflag=1&cursor='
testcom2 = r'&scorecursor=0&orirepnum=2&reporder=o&reppageflag=1&source=1&_=1540974906630'
last = 0
ct = []
ctid = []
ctn = []
tt = []

while 1:
    testcom = testcom1+str(last)+testcom2
    t1 = requests.get(testcom,headers = headers)
    t2 = t1.text
    t3 = re.search('\\((.+)\\)',t2)
    t4 = json.loads(t3.group(1))
    oriretnum = t4['data']['oriretnum']
    #print(oriretnum)
    if not t4['data']['last']:
        print('爬取完毕！')
        break;
    for j in range(oriretnum):
        c_t = t4['data']['oriCommList'][j]['content']
        ct.append(c_t)
        ct_id = t4['data']['oriCommList'][j]['userid']
        ctid.append(ct_id)
        t_t = t4['data']['oriCommList'][j]['time']
        t_t_ = datetime.datetime.fromtimestamp(int(t_t)).strftime('%Y-%m-%d %H:%M:%S')
        tt.append(t_t_)
        ct_n = t4['data']['userList'][str(ct_id)]['nick']
        ctn.append(ct_n)
    last = t4['data']['last']
    #print(last)   
#print(ctid)
#print(ctn)
#print(tt)
```
