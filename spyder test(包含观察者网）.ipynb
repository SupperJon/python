# 观察者网 综合

```python
import re
from bs4 import BeautifulSoup
import requests
import urllib 
import csv
import json

def gethtml(url):
    try:
        headers = {'User_Agent':
                   'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.168 Safari/537.36'}
        r = requests.get(url, timeout = 30,headers = headers)
        r.raise_for_status()
        r.encoding = 'utf-8'
        return r.text
    except:
        return "产生异常"

    
url = input()    
test = gethtml(url)    
comment_id = re.findall(r'data-id=\"(\d*)\"',test)    

#order用作排序 ,order=1为按照最新排序，order=2为按照最热排序，order=3为按照最早排序
url_1 = r'https://user.guancha.cn/comment/cmt-list.json?codeId='
url_2 = r'&codeType=1&pageNo='
url_3 = r'&order=1&ff=www'

baosong_list = []
k = 1
count = 0
total = 1
while count<int(total):
    comment_url = url_1 + comment_id[0]+ url_2+str(k)+url_3    
    #print(comment_url)
    comment_text = gethtml(comment_url)
    js = json.loads(comment_text)
    total = js['count']
    cm_user = js['items'][0]['user_nick']
    cm_comment = js['items'][0]['content']
    k = k+1
    for i in range(len(js['items'])):
        count = count +1
        nick = js['items'][i]['user_nick']
        comment = js['items'][i]['content']
        #print(nick)
        #print(comment)
        baosong = '“观察者网”用户“'+nick+'”：'+comment
        baosong_list.append(baosong)
        #print('第%d条评论：'%count)
        #print(baosong)
        print('“观察者网”用户“%s”：%s'%(nick,comment))
print(count)
print(total)
```
